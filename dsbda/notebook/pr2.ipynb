```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stud_df=pd.read_csv('/Users/shreyuu/VS_Code_projects/junk/6th-sem-practicals/dsbda/datasets/StudentsPerformance.csv')
stud_df.head(20)
```

```python
stud_df.info()
```

```python
stud_df.loc[10:14, "math score"] = np.nan
stud_df.loc[50:59, "reading score"] = np.nan
stud_df.loc[70:79, "writing score"] = np.nan
```

```python
stud_df.head(20)
```

```python
missing_val = stud_df.isnull().sum()
missing_val
```

```python
numeric_cols = ["math score", "reading score", "writing score"]
stud_df[numeric_cols] = stud_df[numeric_cols].astype(float)

stud_df[stud_df[numeric_cols] < 0] = np.nan

stud_df.fillna(stud_df.mean(), inplace=True)

stud_df.head(20)
```

```python
plt.subplot(1,2,1)
sns.histplot(stud_df["math score"], kde=True)
plt.title("math score")
```

```python
plt.subplot(1,2,1)
sns.histplot(stud_df["math score"], kde=True)
plt.title("reading score")

plt.subplot(1,2,2)
sns.histplot(stud_df["math score"], kde=True)
plt.title("writing score")
```

```python
z_scores =(stud_df[numeric_cols] - stud_df[numeric_cols].mean())/stud_df[numeric_cols].std()
outliers = (z_scores > 3) | (z_scores < -3)
stud_df[outliers] = np.nan
```
