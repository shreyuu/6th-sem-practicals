# TITANIC DATASET

```python
import pandas as pd
titanic_df = pd.read_csv("/Users/shreyuu/VS_Code_projects/junk/6th-sem-practicals/dsbda/datasets/titanic.csv")
titanic_df.head(10)
```

```python
grouped_stats = titanic_df.groupby('Pclass')
grouped_stats["Age"].describe()
```

```python
mean_age_by_class = grouped_stats["Age"].mean().tolist()
print("mean age for each passenger class\n", mean_age_by_class)
```

```python
median_age_by_class = grouped_stats["Age"].median().tolist()
print("median age for each passenger class\n", median_age_by_class)
```

```python
min_age_by_class = grouped_stats["Age"].min().tolist()
print("min age for each passenger class\n", min_age_by_class)
```

```python
max_age_by_class = grouped_stats["Age"].max().tolist()
print("max age for each passenger class\n", max_age_by_class)
```

```python
sd_age_by_class = grouped_stats["Age"].std().tolist()
print("Standard Deviation of age for each passenger class\n", sd_age_by_class)
```
